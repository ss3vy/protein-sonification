# Define the Necessary Functions

````python
def dssp_to_df(filename):
    import pandas as pd
    dssp_raw = open(filename, 'r', encoding='utf-8-sig').read()
    dssp_raw = dssp_raw.split('\n')
    dssp_table = dssp_raw[28:]
    cols = ['residue num', 'chain','AA', 'ss']
    rows = []
    for row in dssp_table:
        if len(row)>0:
            rows.append([row[5:10], row[11], row[13], row[16]])
    df = pd.DataFrame.from_records(rows, columns=cols)
    
    return df

def add_musical_ss_col(df):
    df.loc[(df['ss'] == 'H') | (df['ss'] == 'G') | (df['ss'] == 'I'), 'music_ss'] = 'helix'
    df.loc[(df['ss'] == 'B') | (df['ss'] == 'E'), 'music_ss'] = 'sheet'
    df.loc[(df['ss'] != 'H') & (df['ss'].str != 'G') & (df['ss'].str != 'I') & (df['ss'] != 'B') & (df['ss'] != 'E'), 'music_ss'] = 'other'
    
    return df

# Add Musical Features
def add_note_features(df, aa_magenta_scale , ss_to_duration):
    df['magenta_scale'] = df.AA.str.strip().replace(aa_magenta_scale)
    df['duration'] = df.music_ss.str.strip().replace(ss_to_duration)
    
    df.loc[0, 'start_time'] = 0
    df.loc[0, 'end_time'] = 0 + df.loc[0, 'duration']
    
    for i in range(1, len(df)):
        df.loc[i, 'start_time'] = df.loc[i-1, 'end_time']
        df.loc[i, 'end_time'] = df.loc[i-1, 'end_time'] + df.loc[i, 'duration']
        
    return df

def add_scale_features(df, aa_scale,col_name = 'chord_scale'):
    df[col_name] = df.AA.str.strip().replace(aa_scale)    
    return df

def make_note_sequence(df, return_dict=False, export_json=True, json_dir = 'assets/note_sequences/', filename='data'):
    note_lst = []
    for i in range(len(df)):
        note_lst.append({'pitch': df.loc[i, 'magenta_scale'], 'startTime': df.loc[i, 'start_time'], 
                         'endTime': df.loc[i, 'end_time'], 'instrument': 1, 'program': 1})
    if export_json:
        data = {'notes': note_lst, 'totalTime':nCov_spike_df.loc[len(note_lst)-1,'end_time']}
        import json
        with open(json_dir+filename +'_note_sequence.json', 'w') as outfile:
            json.dump(data, outfile)  
    if return_dict:
        return data
    
    
def json_sequence(df, return_dict=False, export_json=True, json_dir = 'assets/sequence_data/', filename='data'):
    lst = []
    for i in range(len(df)):
        lst.append({'chord': df.loc[i, 'aa_chord_tone_js'], 'startTime': df.loc[i, 'start_time'], 
                         'endTime': df.loc[i, 'end_time'], 'amino_acid': df.loc[i, 'AA'], 'duration': df.loc[i, 'duration']})
    import json
    with open(json_dir+filename +'_seq_chords.json', 'w') as outfile:
        json.dump(lst, outfile)  
    if return_dict:
        return data
````

# Static Visual Preparation

````python
import drawSvg as draw
import pandas as pd
````

````python
cmap = {'P': '#000098', 'N': '#0001FF','D': '#0082FF','S': '#00FFFF',
        'Q': '#008200','E': '#00C800','R': '#63E601','M': '#C60100',
        'I': '#E43301','L': '#E46400','F': '#FEFF01','Y': '#5B2500',
        'W': '#6C3C01','K': '#8A5B02','H': '#B19600','A': '#590099',
        'G': '#770099','V': '#B30099','C': '#DB009A','T': '#8B8C8D',
        '!': '#FFF'}
````

````python
# set seconds of playback
n = 60

ACE2 = pd.read_csv('assets/dataframes/ACE2_ssresid.csv', index_col = 'Unnamed: 0')
first_n_ACE2 = ACE2[ACE2['end_time']<=n]

SarsCoV2Spike = pd.read_csv('assets/dataframes/nCov_spike_ssresid.csv', index_col = 'Unnamed: 0')
first_n_spike = SarsCoV2Spike[SarsCoV2Spike['end_time']<=n]
first_n_spike.head()
````

````python
def make_aa_roll(df, scale_factor = 10, fig_width = 300):
    fig_height = scale_factor * df.iloc[-1:,-1:].values[0][0]

    d = draw.Drawing(fig_width, fig_height, displayInline=True)

    for i in range(len(df)-1, -1, -1):
        aa = df.loc[i, 'AA']
        color = cmap[aa]
        h=scale_factor*df.loc[i, 'duration']
        
        # Draw rectangle
        d.append(draw.Rectangle(0, fig_height-h-scale_factor*df.loc[i, 'start_time'], # x, y
                                fig_width, 
                                h, 
                                fill=color,
                                stroke='black',
                                stroke_width = 0.2))
    return d
````
